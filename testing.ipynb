```python
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
```

```python
cnn = tf.keras.models.load_model('C:/Users/yashg/OneDrive/Desktop/Vegetable_Sorting_Algo/train_model.h5')
```

```python
import cv2
image_path = 'C:/Users/yashg/OneDrive/Desktop/Vegetable_Sorting_Algo/test/watermelon/Image_1.jpg'
img = cv2.imread(image_path)
plt.imshow(img)
plt.title('Test Image')
```

```python
image = tf.keras.preprocessing.image.load_img(image_path, target_size=(64,64))
input_arr = tf.keras.preprocessing.image.img_to_array(image)
```

```python
input_arr = np.array([input_arr])
prediction = cnn.predict(input_arr)
```

```python
print(prediction)
```

```python
test_set = tf.keras.utils.image_dataset_from_directory(
    'C:/Users/yashg/OneDrive/Desktop/Vegetable_Sorting_Algo/test',
    labels="inferred",
    label_mode="categorical",
    class_names=None,
    color_mode="rgb",
    batch_size=32,
    image_size=(64,64),
    shuffle=True,
    seed=None,
    validation_split=None,
    subset=None,
    interpolation="bilinear",
    follow_links=False,
    crop_to_aspect_ratio=False
)
```
